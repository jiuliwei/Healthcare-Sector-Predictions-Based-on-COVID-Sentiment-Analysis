# Create New Data File

```python
for index in range(69):
    #data = pd.read_csv("/Users/joye/学习/PA/project/tweetid_userid_keyword_sentiments_emotions_United States.csv")
    data = pd.read_csv("/Users/joye/学习/PA/project/chunk"+str(index)+".csv")
    data = data.drop('tweet_ID', axis=1)
    data = data.drop('user_id', axis=1)
    data = data.drop('keywords', axis=1)
    data.loc[1000000] = data.loc[0]
    df = pd.DataFrame(columns=('tweet_timestamp','count','sent_-2','sent_-1','sent_0','sent_1','sent_2','emo_A',
                              'emo_F','emo_H','emo_S','valence_intensity','fear_intensity','anger_intensity',
                              'happiness_intensity','sadness_intensity'))
    count = 0
    date = data['tweet_timestamp'][0].split(' ')[0]
    row = {'tweet_timestamp':date,'count':0,'sent_-2':0,'sent_-1':0,'sent_0':0,'sent_1':0,'sent_2':0,'emo_A':0,
                              'emo_F':0,'emo_H':0,'emo_S':0,'valence_intensity':0,'fear_intensity':0,'anger_intensity':0,
                              'happiness_intensity':0,'sadness_intensity':0}
    total = 0

    n = 0
    for i in range(len(data)):
        if data['tweet_timestamp'][i].split(' ')[0] != date :
            print(date+":"+str(count))
            row['count'] = count
            row['valence_intensity'] /= count
            row['fear_intensity'] /= count
            row['anger_intensity'] /= count
            row['happiness_intensity'] /= count
            row['sadness_intensity'] /= count
            df.loc[n]=row
            n+=1
            total+=count
            date = data['tweet_timestamp'][i].split(' ')[0]
            count = 0
            row = {'tweet_timestamp':date,'count':0,'sent_-2':0,'sent_-1':0,'sent_0':0,'sent_1':0,'sent_2':0,'emo_A':0,
                              'emo_F':0,'emo_H':0,'emo_S':0,'valence_intensity':0,'fear_intensity':0,'anger_intensity':0,
                              'happiness_intensity':0,'sadness_intensity':0}
        count+=1;
        row['sent_'+str(data['sentiment'][i])] += 1
        if data['emotion'][i]=='A' or data['emotion'][i]=='F'or data['emotion'][i]=='H'or data['emotion'][i]=='S':
            row['emo_'+data['emotion'][i]]+=1
        row['valence_intensity'] += data['valence_intensity'][i]
        row['fear_intensity'] += data['fear_intensity'][i]
        row['anger_intensity'] += data['anger_intensity'][i]
        row['happiness_intensity'] += data['happiness_intensity'][i]
        row['sadness_intensity'] += data['sadness_intensity'][i]

    print(date+":"+str(count))
    total+=count
    print("total: "+str(total))
    df.to_csv('output'+str(index)+'.csv',index=False)
```

```python
df = pd.read_csv("output0.csv")

for i in range(1,69):
    print(i)
    data = pd.read_csv("output"+str(i)+".csv")
    k = len(df)-1
    if df['tweet_timestamp'][k] == data['tweet_timestamp'][0]:
        df['sent_-2'][k] += data['sent_-2'][0]
        df['sent_-1'][k] += data['sent_-1'][0]
        df['sent_0'][k] += data['sent_0'][0]
        df['sent_1'][k] += data['sent_1'][0]
        df['sent_2'][k] += data['sent_2'][0]
        df['emo_A'][k] += data['emo_A'][0]
        df['emo_F'][k] += data['emo_F'][0]
        df['emo_H'][k] += data['emo_H'][0]
        df['emo_S'][k] += data['emo_S'][0]
        count = df['count'][k] + data['count'][0]
        df['valence_intensity'][k] = (df['valence_intensity'][k]*df['count'][k]+
                                          data['valence_intensity'][0]*data['count'][0])/count
        df['fear_intensity'][k] = (df['fear_intensity'][k]*df['count'][k]+
                                          data['fear_intensity'][0]*data['count'][0])/count
        df['anger_intensity'][k] = (df['anger_intensity'][k]*df['count'][k]+
                                          data['anger_intensity'][0]*data['count'][0])/count
        df['happiness_intensity'][k] = (df['happiness_intensity'][k]*df['count'][k]+
                                          data['happiness_intensity'][0]*data['count'][0])/count
        df['sadness_intensity'][k] = (df['sadness_intensity'][k]*df['count'][k]+
                                          data['sadness_intensity'][0]*data['count'][0])/count
        df['count'][k] = count
        data=data.drop(0, axis=0)
    df = pd.concat([df,data],axis = 0)
    df = df.reset_index(drop=True)
```

```python
df.to_csv('output.csv',index=False)
```
